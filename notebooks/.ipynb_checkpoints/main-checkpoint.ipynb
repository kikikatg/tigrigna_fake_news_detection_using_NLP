```python
import pandas as pd

data = pd.read_csv("../datasets/tigrigna_fake_news_combined.csv", encoding="utf-8")

print(data.shape)
data.head()
```

```python
print(data.columns)
print(data['label'].value_counts())
```

```python
data.rename(columns={'Tigrigna_translated_text': 'text'}, inplace=True)
```

```python
import re

def clean_text(text):
    text = str(text)
    
    # Remove URLs
    text = re.sub(r'https?://\S+', '', text)
    
    # Remove Tigrigna punctuation explicitly
    text = re.sub(r'[፡።፣፤፥፦]', '', text)
    
    # Remove English punctuation
    text = re.sub(r'[^\w\sሀ-፿]', '', text)
    
    # Normalize spaces
    text = re.sub(r'\s+', ' ', text)
    
    return text.strip()
    

data['clean_text'] = data['text'].apply(clean_text)
```

```python
def normalize_tigrigna(text):
    text = text.replace("አ", "ኣ")
    text = text.replace("ሐ", "ሀ")
    return text

data['clean_text'] = data['clean_text'].apply(normalize_tigrigna)
```

```python
def tokenize(text):
    tokens = re.findall(r'[A-Za-z]+|[ሀ-፿]+', text)
    return [word.strip() for word in tokens]

data['tokens'] = data['clean_text'].apply(tokenize)
```

```python
stopwords = set([
    "እዚ","እዩ","እቲ","ኣብ","ናብ","ን","ናይ",
    "ከም","ስለ","ካብ","ወይ","ምስ","ነቲ","ነዚ",
    "ኣብዚ","ኝ","ይኹን","እምበር","ዋላኳ",
    "ኮታስ","ብ","እዉን","ድማ","ከሎ","እዋን",
    "ነታ","እየ","እዮም","እውን","ግን","ከምዘሎ","ኣሎ"
])
```

```python
def remove_affixes(word, stopwords):
    for sw in stopwords:
        if word.startswith(sw) and len(word) > len(sw):
            word = word[len(sw):]

        if word.endswith(sw) and len(word) > len(sw):
            word = word[:-len(sw)]

    return word
```

```python
def clean_tokens(tokens):
    cleaned = []

    for word in tokens:
        w = word.strip()

        # Remove prefix/suffix stopwords
        w = remove_affixes(w, stopwords)

        # Remove if empty or still a stopword
        if w == "" or w in stopwords:
            continue

        cleaned.append(w)

    return cleaned
```

```python
data['tokens'] = data['tokens'].apply(clean_tokens)
```

```python
data['processed_text'] = data['tokens'].apply(lambda x: " ".join(x))
```
